```python
from pyspark.sql import SparkSession
import os

# 1. Caminhos absolutos
current_dir = os.path.abspath(".")
warehouse_path = os.path.join(current_dir, "spark-warehouse")
metastore_path = os.path.join(current_dir, "metastore_db")

for path in [warehouse_path, metastore_path]:
    if not os.path.exists(path):
        os.makedirs(path)
        print(f"Pasta criada: {path}")

# 3. Configuração do Spark
spark = (SparkSession.builder \
    .appName("CaseAmbev") \
    .config("spark.jars.packages", "io.delta:delta-spark_2.12:3.1.0") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .config("spark.sql.catalog.abinbev", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .config("spark.sql.warehouse.dir", warehouse_path) \
    .config("javax.jdo.option.ConnectionURL", f"jdbc:derby:;databaseName={metastore_path};create=true") \
    .config("javax.jdo.option.ConnectionDriverName", "org.apache.derby.jdbc.EmbeddedDriver") \
    .config("spark.driver.extraJavaOptions", f"-Dderby.system.home={current_dir} -Dderby.stream.error.file={current_dir}/derby.log") \
    .enableHiveSupport() \
    .getOrCreate())
```
