```python
comments = {
    "DATE": "Data original da transação de venda extraída do arquivo",
    "ce_brand_flvr": "Código identificador único da marca e sabor do produto",
    "BRAND_NM": "Nome descritivo da marca do produto",
    "Btlr_Org_LVL_C_Desc": "Descrição da região ou unidade organizacional do engarrafador",
    "CHNL_GROUP": "Grupo de canal de venda de alto nível",
    "TRADE_CHNL_DESC": "Descrição detalhada do canal de comercialização",
    "PKG_CAT": "Código da categoria de embalagem",
    "Pkg_Cat_Desc": "Descrição completa da categoria da embalagem",
    "TSR_PCKG_NM": "Nome comercial da embalagem/pacote",
    "volume": "Volume total comercializado (valor sanitizado)",
    "YEAR": "Ano de referência da venda",
    "MONTH": "Mês de referência da venda",
    "PERIOD": "Período/Trimestre fiscal da transação",
    "dtload": "Data de processamento e carga no Lakehouse (Chave de Partição)"
}

for column, comment in comments.items():
    spark.sql(f"ALTER TABLE delta.`{bronze}sales` ALTER COLUMN {column} COMMENT '{comment}'")

spark.sql(f"describe delta.`{bronze}sales`").toPandas().style.hide(axis='index')
```
