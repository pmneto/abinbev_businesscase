```python
from pyspark.sql.functions import col, upper, trim 
from pyspark.sql.types import StructType , DateType, StringType
```

```python
df = spark.read.format('delta').load(f'{bronze}channel_group')
```

```python
df = (
    df
    .withColumnRenamed('TRADE_CHNL_DESC','tpchannel')
    .withColumnRenamed('TRADE_GROUP_DESC','tpgroup')
    .withColumnRenamed('TRADE_TYPE_DESC','tptrade')
     )
```

```python
strings = [i.name for i in df.schema if isinstance(i.dataType, StringType) == True]
```

```python
for column in df.columns:

    if column in strings:
     df = (df.withColumn(column, upper(trim(col(column)))))
    
```

```python
df = df.dropDuplicates()
```
